```python
import random
```

```python
income = []
age = []
```

```python
for i in range(50):
    a = random.randrange(18, 65)
    age.append(a)
    if a <= 30:
        income.append(random.randint(2000, 3500))
    if a > 30 and a <= 55:
        income.append(random.randint(3200, 5100))
    if a > 55:
        income.append(random.randint(4850, 6000))
```

```python
plt.scatter(income, age)
plt.show()
```

```python
age_income = pd.DataFrame({'age': age, 'income': income})
```

```python
sorted_data = age_income.sort_values(by=['age'])
```

```python
sorted_data['income_age_average'] = sorted_data.income.rolling(8, min_periods=2, win_type='gaussian').mean(std=1)
```

```python
sorted_age = sorted_data.age
sorted_income = sorted_data.income
sorted_avg = sorted_data.income_age_average
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,1,2)
plt.scatter(sorted_age, sorted_income)
plt.twinx()
plt.plot(sorted_age, sorted_avg, color='orange')
plt.show()
```
